```python
from sympy import *
```

```python
x,y = symbols('x,y')
```

```python
f = sqrt(1-x**2-y**2)
f
```

```python
plotting.plot3d(f,(x,-1,1),(y,-1,1))
```

```python
plotting.plot_contour(f,(x,-1,1),(y,-1,1))
```

```python
limit(sin(x**2)/x**2,x,0)
```

```python
series(sin(x**2),x,x0=0,n=10)
```
