```python
from sympy import *
```

## Example on page 30

```python
x,y = symbols('x,y')
```

```python
f = 1/(1+x**2+y**2)
```

```python
plotting.plot3d(f,(x,-3,3),(y,-3,3))
```

```python
plotting.plot_contour(f,(x,-3,3),(y,-3,3))
```

```python
diff(f,x)
```

```python
diff(f,y)
```

```python
# plot of tangent plane at x0 = 1/2, y0 = 1
t = f.subs({x:0.5, y:1}) + diff(f,x).subs({x:0.5, y:1}) * (x-0.5) + diff(f,y).subs({x:0.5, y:1}) * (y-1)
t
```

```python
plotting.plot3d(f,t,(x,-3,3),(y,-3,3))
```
